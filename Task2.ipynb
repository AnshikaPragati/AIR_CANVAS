```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

Gray_image= cv2.imread("/content/bulb.jpeg",0)
retr,binary= cv2.threshold(Gray_image,200,255,cv2.THRESH_BINARY)

contours,hierarchy= cv2.findContours(binary,cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE)
image1_copy= cv2.imread("/content/bulb.jpeg")
cv2.drawContours(image1_copy,contours,-1, (0,0,255),thickness=2, lineType =cv2.LINE_AA)

plt.figure(figsize=[10,10])
plt.imshow(image1_copy[:,:,::-1]);plt.axis('off');

```
